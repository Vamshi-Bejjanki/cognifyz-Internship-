# Required Imports for level-1

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## TASK-1 : Top Cuisines

### Determine the top three most common cuisines in the dataset.

```python
data=pd.read_csv("E:\\congnifyz_internship\\Dataset .csv")
top_cui = data['Cuisines'].value_counts().head(3)
print("Top 3 most common cuisines:-\n",top_cui)
plt.figure(figsize=(6, 4))
sns.barplot(x=top_cui.index, y=top_cui.values, hue=top_cui.index, palette='viridis', legend=False)
plt.title('Top 3 Most Common Cuisines')
plt.xlabel('Cuisines')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

### Calculate the percentage of restaurants that serve each of the top cuisines.

```python
tot_cui = data['Cuisines'].value_counts()
per_res=(tot_cui / len(data))*100
print("Percentage of cuisines:-")
print(per_res)
per_top_res=(top_cui / len(data))*100
print("\nPercentage of top cuisines:-")
print(per_top_res)
plt.figure(figsize=(6, 4))
sns.barplot(x=top_cui.index, y=per_top_res.values, hue=top_cui.index, palette='viridis', legend=False)
plt.title('Percentage of Top 3 Cuisines')
plt.xlabel('Cuisines')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## TASK-2: City Analysis

### Identify the city with the highest number of restaurants in the dataset.

```python
city_co = data['City'].value_counts()
city_co = data['City'].value_counts()
top_city = city_co.idxmax()
top_city_co = city_co.max()
print(f"The city with the highest number of restaurants is {top_city} with {top_city_co} restaurants.")
```

### Calculate the average rating for restaurants in each city.

```python
city_avg_rtg = data.groupby('City')['Aggregate rating'].mean()
print("Average ratings for restaurants in each city:")
print(city_avg_rtg)
```

### Determine the city with the highest average rating.

```python
top_city_avg_rtg = city_avg_rtg.max()
print("City name :->",top_city_co)
print("Average Ratings:->",top_city_avg_rtg)
```

## TASK-3: Price Range Distribution

### Create a histogram or bar chart to visualize the distribution of price ranges among the restaurants.

```python
print("\n")

plt.figure(figsize=(6, 4)) 
data['Price range'].value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.title(' Price Ranges Distribution')
plt.xlabel('Price Range')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=0)
plt.show()
```

### Calculate the percentage of restaurants in each price range categoryn

```python
pri_rge_co = data['Price range'].value_counts()
tot_res = len(data)
pri_rge_per = (pri_rge_co / tot_res) * 100
print("Percentage of restaurants in each price range:")
print(pri_rge_per)
plt.figure(figsize=(6, 4))
sns.barplot(x=pri_rge_per.index, y=pri_rge_per.values, hue=pri_rge_per.index, palette='viridis', legend=False)
plt.title('Percentage of Restaurants in Each Price Range')
plt.xlabel('Price Range')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## TASK-4: Online Delivery

### Determine the percentage of restaurants that offer online delivery.

```python
data['Has Online delivery'] = data['Has Online delivery'].apply(lambda x: 1 if x == 'Yes' else 0)
delivery_available = data['Has Online delivery'].sum()
total_restaurants_count = data.shape[0]
percentage_online_delivery = (delivery_available / total_restaurants_count) * 100
print(f"Percentage of Restaurants Offering Online Delivery: {percentage_online_delivery:.2f}%")
```

### Compare the average ratings of restaurants with and without online delivery

```python
delivery_ratings = data[data['Has Online delivery'] == 'Yes']['Aggregate rating']
no_delivery_ratings = data[data['Has Online delivery'] == 'No']['Aggregate rating']
avg_delivery = delivery_ratings.mean()
avg_no_delivery = no_delivery_ratings.mean()
print(f"Average rating for restaurants with online delivery: {avg_delivery:.2f}")
print(f"Average rating for restaurants without online delivery: {avg_no_delivery:.2f}\n")
categories = ['With Online Delivery', 'Without Online Delivery']
averages = [avg_delivery, avg_no_delivery]

plt.bar(categories, averages, color=['blue', 'orange'])
plt.ylabel('Average Rating')
plt.title('Comparison of Average Ratings')
plt.ylim(0, 5)  
plt.show()
```
